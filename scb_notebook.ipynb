```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import scb_salaries as scbs
import scb_example as scbe
```

```python
df = scbs.get_dataframe(scbe.salaryset)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(30,20))
fig.suptitle("Income distribution", fontsize=24)
fig.patch.set_facecolor('white')

sns.barplot(x='age', y='median', hue='gender', data=df, ax=ax1)
sns.barplot(x='age', y='average', hue='gender', data=df, ax=ax2)

# ax1.set_title("Title for first  plot")
# ax2.set_title("Title for second plot")
```
